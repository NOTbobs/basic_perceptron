```python
import numpy as np
import csv
import random
import matplotlib.pyplot as plt
```

```python
#process data
rawdata=[]
data=[]
with open('iris.data', newline='') as csvfile:
     spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
     for row in spamreader:
         rawdata.append(row)
rawdata=np.array(rawdata)

for i in range(len(rawdata)): 
  if rawdata[i][4]=='Iris-setosa': 
    rawdata[i][4]=1
  else: 
    rawdata[i][4]=-1

rawdata=rawdata.astype('float32')
#parse data and labels
data=rawdata[:,range(0,4)]
labels=rawdata[:,4]
```

```python
def perceptron(inputs,weights): 
  return np.sign(np.dot(weights.T,inputs))


def update(inputs,labels,weights,lr=0.001): 
  out=np.sign(perceptron(inputs,weights))
  #print(out)
  if out==1.0 and labels==-1.0: #decrease
    weights-=lr*np.average(weights*inputs,axis=0).reshape(4,1)
    #print('updating....')
  elif out==-1.0 and labels==1.0: #increase:
    weights+=lr*np.average(weights*inputs,axis=0).reshape(4,1)
    #print('updating....')
  else:
    pass
    #print('correct')
  return weights
```

```python
random.seed(1001)
weights=np.ones((4,1))
#to randomize dataset
randomization=random.sample(range(0,150),150)
```

```python
validation_accuracy=0
validation_graph=[]
for epochs in range(10):
  for i in range(100): 
    weights=update(data[randomization[i]],labels[randomization[i]],weights,lr=0.01)
    for validation in range(100,150):
      if perceptron(data[randomization[validation]],weights) == labels[randomization[validation]]:
        validation_accuracy+=1
    print("validation :", (validation_accuracy/50)*100, '%')
    validation_graph.append((validation_accuracy/50)*100)
    validation_accuracy=0    
```

```python
#validation_accuracy versus epochs. 
plt.plot(validation_graph)
```
